```python
# binary Thesholding
img = cv2.imread('4.png', cv2.IMREAD_GRAYSCALE)
thresh = 0
maxvalue = 255

th, img2 = cv2.threshold(img, thresh, maxvalue, cv2.THRESH_BINARY)

# cv2.imshow('origional image', img)
# cv2.imshow('threshold image', img2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# dilation
dilationSize = 6
element = cv2.getStructuringElement(cv2.MORPH_CROSS, (2*dilationSize+1, 2*dilationSize+1), (dilationSize, dilationSize))
imgdilated = cv2.dilate(img, element)

# erosion
erosionSize = 6
element = cv2.getStructuringElement(cv2.MORPH_CROSS, (2*erosionSize+1, 2*erosionSize+1), (erosionSize, erosionSize))
imgerosion = cv2.erode(img, element)

# Morphological Opening Operation
img = cv2.imread('4.png', cv2.IMREAD_GRAYSCALE)
openingSize = 6
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*openingSize+1, 2*openingSize+1), (openingSize, openingSize))
imageMorphOpened = cv2.morphologyEx(img, cv2.MORPH_OPEN, element, iterations= 3 )

# Morphological Closing Operation
img = cv2.imread('4.png', cv2.IMREAD_GRAYSCALE)
closingSize = 6
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*closingSize+1, 2*closingSize+1), (closingSize, closingSize))
imageMorphOpened = cv2.morphologyEx(img, cv2.MORPH_CLOSE, element, iterations= 3 )

# cv2.namedWindow('img', cv2.WINDOW_NORMAL)
# cv2.imshow('img', imageMorphOpened)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
def displayConnectedComponents (img) :
    imlabels = img
    # The following line finds the min and max pixel values
    (minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(imlabels)
    # Normalize the imageso the min value is 0 and max value is 255.
    imlabels = 255 * (imlabels - minVal)/(maxVal-minVal)
    # Convert image to 8-bits unsigned type
    imlabels = np.uint8(imlabels)
    # Apply colormapped labels
    imcolor = cv2.applyColorMap(imlabels, cv2.COLORMAP_JET)
    # Display colormapped labels
    cv2.namedWindow('img', cv2.WINDOW_NORMAL)
    cv2.imshow('img', imcolor)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

img = cv2.imread('4.png', cv2.IMREAD_GRAYSCALE)
th, imthresh = cv2.threshold(img, 127 , 255 , cv2.THRESH_BINARY)
# Find connected components
_, imlabels = cv2.connectedComponents(imthresh)
displayConnectedComponents(imlabels)
```

```python
# Geometric Tranformations

# Rotation 
img = cv2.imread('1.jpeg')
dim = img.shape

rotation = -30
scalefactor = 1

rotationmatrix = cv2.getRotationMatrix2D((dim[1]/2, dim[0]/2), rotation, scalefactor)
rotation_img = cv2.warpAffine(img, rotationmatrix , (dim[1], dim[0]))

# Affine
warpmat1 = np.float32([[1.2, 0.2,2], [-0.3,1.3,1]])

result = cv2.warpAffine(img, warpmat1,(int(1.5*dim[0]), int(1.4*dim[1])), flags = cv2.INTER_LINEAR, borderMode = cv2.BORDER_REFLECT_101)

cv2.namedWindow('img', cv2.WINDOW_NORMAL)
cv2.imshow('img', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
